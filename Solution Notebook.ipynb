## Introduction to AI and ML

In this chapter, we'll take a look at autonomous and intelligent agents, machine learning approaches, and the problem of classification.

### Artificial Intelligence

Artificial intelligence (AI) is a branch of computer science dealing with the simulation of intelligent behavior in computers. It is the capability of a machine to imitate intelligent human behavior. 

AI is the study of Intelligent Agents (IA). An **intelligent agent** is an autonomous entity that can:
* Perceive and act in its environment using sensors and actuators, respectively.
* Direct its activity toward goals and use knowledge and learning in achieving those goals.
* Analyze itself in terms of behavior, error, and success and adapt online, possibly in real time.
* Be organized with other agents, typically in a hierarchy to tackle problems at different abstraction levels.

Throughout this notebook, we will refer to intelligent agents as **bots** or **AIs**.

### Types of Intelligent Agents

#### Simple Reflex Agents

Simple reflex agents act only on the basis of the current percept, ignoring the rest of the percept history. The agent function is based on the condition-action rule: "if condition, then action". This agent function only succeeds when the environment is fully observable. Some reflex agents can also contain information on their current state which allows them to disregard conditions whose actuators are already triggered. Infinite loops are often unavoidable for simple reflex agents operating in partially observable environments. Note: If the agent can randomize its actions, it may be possible to escape from infinite loops.

<img src="Simple_reflex_agent.png" width="600">

#### Model-Based Agents

A model-based agent can handle partially observable environments. Its current state is stored inside the agent maintaining some kind of structure which describes the part of the world which cannot be seen. This knowledge about "how the world works" is called a model of the world, hence the name "model-based agent". A model-based reflex agent should maintain some sort of internal model that depends on the percept history and thereby reflects at least some of the unobserved aspects of the current state. Percept history and impact of action on the environment can be determined by using internal model. It then chooses an action in the same way as reflex agent. An agent may also use models to describe and predict the behaviors of other agents in the environment.

<img src="Model_based_reflex_agent.png" width="600">

#### Goal-Based Agents

Goal-based agents further expand on the capabilities of the model-based agents, by using "goal" information. Goal information describes situations that are desirable. This allows the agent a way to choose among multiple possibilities, selecting the one which reaches a goal state. Search and planning are the subfields of artificial intelligence devoted to finding action sequences that achieve the agent's goals.

<img src="Model_based_goal_based_agent.png" width="600">

#### Utility-Based Agents

Goal-based agents only distinguish between goal states and non-goal states. It is possible to define a measure of how desirable a particular state is. This measure can be obtained through the use of a utility function which maps a state to a measure of the utility of the state. A more general performance measure should allow a comparison of different world states according to exactly how happy they would make the agent. The term utility can be used to describe how "happy" the agent is.

A rational utility-based agent chooses the action that maximizes the expected utility of the action outcomes - that is, what the agent expects to derive, on average, given the probabilities and utilities of each outcome. A utility-based agent has to model and keep track of its environment, tasks that have involved a great deal of research on perception, representation, reasoning, and learning.

<img src="Model_based_utility_based.png" width="600">

#### Learning Agents

Learning has the advantage that it allows the agents to initially operate in unknown environments and to become more competent than its initial knowledge alone might allow. A "learning element" is responsible for making improvements, and a "performance element" is responsible for selecting external actions. The learning element uses feedback from the "critic" on how the agent is doing and determines how the performance element should be modified to do better in the future. The performance element is what we have previously considered to be the entire agent: it takes in percepts and decides on actions. The last component of the learning agent is the "problem generator". It is responsible for suggesting actions that will lead to new and informative experiences.

<img src="IntelligentAgent-Learning.png" width="600">

### Machine Learning

Machine Learning (ML) is a subdomain of AI and is the science of getting computers to act without being explicitly programmed. If a program can improve how it performs a certain task based on previous experience then you can say that the program has learnt. This can also be expressed as follows:

If the performance ***P*** of a computer program at completing a task ***T*** improves with experience ***E***, the program is said to ***learn*** from experience ***E***.

### ML Categories

#### Supervised Learning

Supervised learning is the machine learning task of learning a function that maps an input to an output based on example input-output pairs. It infers a function from labeled training data consisting of a set of training examples. In supervised learning, each example is a pair consisting of an input object (typically a vector) and a desired output value (also called the supervisory signal). A supervised learning algorithm analyzes the training data and produces an inferred function, which can be used for mapping new examples. An optimal scenario will allow for the algorithm to correctly determine the class labels for unseen instances. This requires the learning algorithm to generalize from the training data to unseen situations in a "reasonable" way.

Supervised learning thrives on well-labeled data. The correct answer is known for each instance of data. The learning algorithm attempts to extract useful patterns and distinguishing features among the training data.

<img src="Supervised.png" width="650">

#### Unsupervised Learning

Unsupervised learning is a term associated with learning without a teacher, also known as self-organization and a method of modelling the probability density of inputs. Cluster analysis (or clustering) is a branch of machine learning that groups the data that has not been labelled, classified or categorized. Instead of responding to feedback, cluster analysis identifies commonalities in the data and reacts based on the presence or absence of such commonalities in each new piece of data.

<img src="Unsupervised.png" width="425">

#### Reinforcement Learning

Reinforcement learning (RL) is an area of machine learning concerned with how software agents ought to take actions in an environment so as to maximize some notion of cumulative reward. RL differs from supervised learning in that labelled input/output pairs need not be presented, and sub-optimal actions need not be explicitly corrected. Instead the focus is finding a balance between exploration (of uncharted territory) and exploitation (of current knowledge).

RL focuses on generating data through actions, rewarding or punishing the system based on outcome, and on learning by trial and error.

<img src="Reinforcement.png" width="750">

### ML Approaches

#### Decision Trees

A decision tree is a decision support tool that uses a tree-like model of decisions and their possible consequences, including chance event outcomes, resource costs, and utility. It is one way to display an algorithm that only contains conditional control statements. Decision tree learning uses a decision tree (as a predictive model) to go from observations about an item (represented in the branches) to conclusions about the item's target value (represented in the leaves). It is one of the predictive modeling approaches used in statistics, data mining and machine learning. Tree models where the target variable can take a discrete set of values are called classification trees; in these tree structures, leaves represent class labels and branches represent conjunctions of features that lead to those class labels. Decision trees where the target variable can take continuous values (typically real numbers) are called regression trees.

<img src="Decision_tree_salary.png" width="600">

#### Neural Networks

Artificial neural networks (ANN) or connectionist systems are computing systems vaguely inspired by the biological neural networks that constitute animal brains. The neural network itself is not an algorithm, but rather a framework for many different machine learning algorithms to work together and process complex data inputs. Such systems "learn" to perform tasks by considering examples, generally without being programmed with any task-specific rules. For example, in image recognition, they might learn to identify images that contain cats by analyzing example images that have been manually labeled as "cat" or "no cat" and using the results to identify cats in other images. They do this without any prior knowledge about cats, for example, that they have fur, tails, whiskers and cat-like faces. Instead, they automatically generate identifying characteristics from the learning material that they process.

<img src="NeuralNetwork.png" width="600">

#### Bayesian (Belief) Networks

A Bayesian network, Bayes network, belief network, decision network, Bayes(ian) model or probabilistic directed acyclic graphical model is a probabilistic graphical model (a type of statistical model) that represents a set of variables and their conditional dependencies via a directed acyclic graph (DAG). Bayesian networks are ideal for taking an event that occurred and predicting the likelihood that any one of several possible known causes was the contributing factor. For example, a Bayesian network could represent the probabilistic relationships between diseases and symptoms. Given symptoms, the network can be used to compute the probabilities of the presence of various diseases.

<img src="BayesianNetworkV2.jpg" width="600">

#### K-Means Clustering

K-means clustering is a method of vector quantization, originally from signal processing, that is popular for cluster analysis in data mining. K-means clustering aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean, serving as a prototype of the cluster. This results in a partitioning of the data space into groups.

<img src="KMeans.png" width="400">

### Classification

In supervised learning, classification is a process related to categorization, the process in which ideas and objects are recognized, differentiated, and understood. Below we can see an example of two ***decision boundaries***. In a statistical-classification problem with two classes, a ***decision boundary*** or ***decision surface*** is a hypersurface that partitions the underlying vector space into two sets, one for each class. A ***classifier*** will classify all the points on one side of the decision boundary as belonging to one class and all those on the other side as belonging to the other class.

<img src="Classification_problem.png" width="400">

When applying ML to a classification problem, a collection of records (called a **training set**) is provided as input to the ML algorithm. Each record contains a set of attributes, one of which is the actual category or class (the right answer). The attributes that may be used as inputs to the learning algorithm are referred to as **features**. The "correct" answer is referred to as the **class**. Consider the following example classification problem:

<img src="Classification01.png" width="600">

A **confusion matrix** may be used to evaluate the performance of an ML model. A confusion matrix, also known as an error matrix, is a specific table layout that allows visualization of the performance of an algorithm, typically a supervised learning one. Each row of the matrix represents the instances in a predicted class while each column represents the instances in an actual class (or vice versa). The name stems from the fact that it makes it easy to see if the system is confusing two classes (i.e. commonly mislabeling one as another).

<img src="ConfusionMatrix.png" width="800">

While accuracy and error rate provide some information on the performance of a model, **precision**, **recall**, and **F1-score** should typically be used instead. In a case where there are many "negative" data points and very few "positive" points, it is entirely possible to have spectacular accuracy (e.g. 99%) with 0% precision on detecting positive samples and 0% recall of positive samples.

## Artificial Neural Networks

As we mentioned in <a href="#Neural-Networks">Section 1.5.2</a>, artificial neural networks (ANNs) are loosely inspired by biology. Neural networks are comprised of layers of connected artificial neurons. Another name for a single neuron is a **perceptron**.

In machine learning, the perceptron is an algorithm for supervised learning of binary classifiers. A binary classifier is a function which can decide whether or not an input, represented by a vector of numbers, belongs to some specific class. It is a type of linear classifier, i.e. a classification algorithm that makes its predictions based on a linear predictor function combining a set of weights with the feature vector.

<img src="Perceptron.png" width="600">

A single perceptron receives one or more inputs, computes the sum of all products of each input against a weight value for each input, and applies an activation function onto the result, producing an output. The multiplied weights are "trainable" parameters, meaning that we can adjust each weight to increase or decrease the importance of each "feature", or "input". 

***A perceptron learning algorithms works as follows:***
1. Randomly initialize the weight vector
2. Compute a loss value (simply put, the difference between the expected vs. actual outputs)
3. Automatically adjust the weights based on a configurable learning rate parameter and the loss value for each weight in an attempt to minimize the loss value.
4. Repeat steps 2-3 until the error is under a specified threshold, or until the number of iterations reaches a pre-defined target.

While perceptrons work well in situations where data points are linearly separable, they are not able to represent more complex non-linear functions, and as such, are not able to handle complex scenarios. A *neural network* uses many layers of linked Perceptrons, where the output of one layer of Perceptrons is used as input into another layer of Perceptrons, and so on.

***A neural network learning algorithms works as follows:***
1. Randomly initialize the weight vectors
2. Forward propagation: Execute the network on a set of examples, computing a loss value (the difference between the expected vs. actual outputs)
3. Back-propagation: Automatically adjust the weights based on a configurable learning rate parameter and error derivatives (with respect to each weight) in an attempt to minimize the loss value.
4. Repeat steps 2-4 until the error is under a specified threshold, or until the number of iterations reaches a pre-defined target.

### Types of Neural Networks

There are many types of neural networks. In fact, research is moving so fast in the field of ML that it is difficult to keep up with new types of networks, and improved network variants to existing known network types. We'll cover only a few basic types of networks, but there are many more!

***Feed-forward Network***

The feedforward neural network was the first and simplest type of artificial neural network devised. In this network, the information moves in only one direction, forward, from the input nodes, through the hidden nodes (if any) and to the output nodes. There are no cycles or loops in the network.

<img src="FeedForward.png">

***Recurrent Network***

A recurrent neural network (RNN) is a class of artificial neural network where connections between nodes form a directed graph along a temporal sequence. This allows it to exhibit temporal dynamic behavior. Unlike feedforward neural networks, RNNs can use their internal state (memory) to process sequences of inputs. This makes them applicable to tasks such as unsegmented, connected handwriting recognition or speech recognition.

<img src="RecurrentNetwork.png">

***Long Short-Term Memory***

Long short-term memory (LSTM) is an artificial recurrent neural network (RNN) architecture used in the field of deep learning. Unlike standard feedforward neural networks, LSTM has feedback connections that make it a "general purpose computer" (that is, it can compute anything that a Turing machine can). It can not only process single data points (such as images), but also entire sequences of data (such as speech or video). For example, LSTM is applicable to tasks such as unsegmented, connected handwriting recognition or speech recognition. Bloomberg Business Week wrote: "These powers make LSTM arguably the most commercial AI achievement, used for everything from predicting diseases to composing music."

<img src="LSTMv2.png" width="500">

***Convolutional Network***

In deep learning, a convolutional neural network (CNN, or ConvNet) is a class of deep neural networks, most commonly applied to analyzing visual imagery. CNNs use a variation of multilayer perceptrons designed to require minimal preprocessing. They are also known as shift invariant or space invariant artificial neural networks (SIANN), based on their shared-weights architecture and translation invariance characteristics.

Convolutional networks were inspired by biological processes in that the connectivity pattern between neurons resembles the organization of the animal visual cortex. Individual cortical neurons respond to stimuli only in a restricted region of the visual field known as the receptive field. The receptive fields of different neurons partially overlap such that they cover the entire visual field.

CNNs use relatively little pre-processing compared to other image classification algorithms. This means that the network learns the filters that in traditional algorithms were hand-engineered. This independence from prior knowledge and human effort in feature design is a major advantage. They have applications in image and video recognition, recommender systems, image classification, medical image analysis, and natural language processing.

<img src="CNN.png">

***Deep Learning***

Deep learning (also known as deep structured learning or hierarchical learning) is part of a broader family of machine learning methods based on learning data representations, as opposed to task-specific algorithms. Learning can be supervised, semi-supervised or unsupervised. 

A deep neural network (DNN) is an artificial neural network (ANN) with multiple layers between the input and output layers. The DNN finds the correct mathematical manipulation to turn the input into the output, whether it be a linear relationship or a non-linear relationship. The network moves through the layers calculating the probability of each output. For example, a DNN that is trained to recognize dog breeds will go over the given image and calculate the probability that the dog in the image is a certain breed. The user can review the results and select which probabilities the network should display (above a certain threshold, etc.) and return the proposed label. Each mathematical manipulation as such is considered a layer, and complex DNN have many layers, hence the name "deep" networks. The goal is that eventually, the network will be trained to decompose an image into features, identify trends that exist across all samples and classify new images by their similarities without requiring human input.

#### Create a model using Keras

The Keras API is the preferred way to create models and layers. This makes it easy to build models and experiment while Keras handles the complexity of connecting everything together.

The Sequential model is a linear stack of layers. In this case, we add two [Dense](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense) layers with 10 and 8 nodes respectively, and an output layer with 3 nodes representing our label predictions. The first layer's `input_shape` parameter corresponds to the number of features from the dataset, and is required.

```python
def create_model():
    model = tf.keras.Sequential()

    model.add(tf.keras.layers.Dense(10,input_shape=(4,),activation='relu'))
    model.add(tf.keras.layers.Dense(8,activation='relu'))
    model.add(tf.keras.layers.Dense(3,activation='softmax'))

    model.compile(tf.keras.optimizers.Adam(learning_rate=0.04),'categorical_crossentropy',metrics=['accuracy'])

    model.summary()
    
    return model
```

As a rule of thumb, increasing the number of hidden layers and neurons typically creates a more powerful model, which requires more data to train effectively.

Note that we use a “softmax” activation function in the output layer. This is to ensure the output values are in the range of 0 and 1 and may be used as predicted probabilities. Finally, the network uses the efficient Adam gradient descent optimization algorithm with a logarithmic loss function, which is called “categorical_crossentropy” in Keras.

### Train the Model

*[Training](https://developers.google.com/machine-learning/crash-course/glossary#training)* is the stage of machine learning when the model is gradually optimized, or the model *learns* the dataset. The goal is to learn enough about the structure of the training dataset to make predictions about unseen data. If you learn *too much* about the training dataset, then the predictions only work for the data it has seen and will not be generalizable. This problem is called *[overfitting](https://developers.google.com/machine-learning/crash-course/glossary#overfitting)*—it's like memorizing the answers instead of understanding how to solve a problem.

The Iris classification problem is an example of *[supervised machine learning](https://developers.google.com/machine-learning/glossary/#supervised_machine_learning)*: the model is trained from examples that contain labels. In *[unsupervised machine learning](https://developers.google.com/machine-learning/glossary/#unsupervised_machine_learning)*, the examples don't contain labels. Instead, the model typically finds patterns among the features.

We can also pass arguments when we call the `fit` function on the model to control the paramters internally used to train the neural network. In the code snippet below, we pass the number of epochs as 200 and batch size as 5 to use when training the model. Debugging is also turned off when training by setting verbose to 0.

A **batch** is a set of N samples. The samples in a batch are processed independently, in parallel. If training, a batch results in only one update to the model. A batch generally approximates the distribution of the input data better than a single input. The larger the batch, the better the approximation; however, it is also true that the batch will take longer to process and will still result in only one update. For inference (evaluate/predict), it is recommended to pick a batch size that is as large as you can afford without going out of memory (since larger batches will usually result in faster evaluation/prediction).

The **epochs** parameter represent an arbitrary training cutoff. An **epoch** is generally defined as "one pass over the entire dataset", and is used to separate training into distinct phases, which is useful for logging and periodic evaluation.

## Neural Network Text Generation

We can also train neural networks to generate text. In this section we'll train a LSTM recurrent neural network to generate sentences.

For this exercise, we'll attempt to generate text similar to what is written in the classic *Alice in Wonderland* text. We'll use a small excerpt of the *Alice in Wonderland* text as input to the training process. The text was downloaded from <a href="http://www.gutenberg.org/">Project Gutenberg</a>, which contains freely downloadable eBooks that are usable for machine learning experiments. This exercise was inspired by <a href="https://machinelearningmastery.com/text-generation-lstm-recurrent-neural-networks-python-keras/">this article</a> written by Dr. Jason Brownlee.

We'll walk through the training code step by step. First, let's import everything we need.

```python
import io
import random
import sys
import json

import numpy as np

import tensorflow as tf
from tensorflow import keras
```

Next, we'll read text from *Alice in Wonderland* (contained in `wonderland.txt`). We'll pre-process the text by combining it all into a single line, and converting it all to lowercase. Finally, we'll break the text up into words by splitting on a ' ' delimiter. We also want to figure out what the unique words are.

```python
# Open the wonderland.txt file, and read the lines of the file.
with io.open('wonderland.txt', encoding='utf-8') as f:
    lines = f.readlines()

# Reconstruct the text by joining each of the lines, inserting a newline character in between each.
text = "\n".join(lines)
print('corpus length:', len(text))

# Convert the text into lowercase, and replace all new line characters with a single empty space.
text = text.lower().replace('\n', ' ')

# Split the string on all spaces. The result is a list of all the words.
words = text.strip().split(' ')

# Add the empty string to the list of words. We can use this string for padding purposes later on when we 
# prepare the input for the neural net algorithms.
words.append('')

# Convert the list of words to a set (to remove duplicates), and back to a list.
unique_words = list(set(words))
print('total unique words:', len(unique_words))
```

Next, we'll create two maps (one from a word to an integer-value, and one from the integer-value back to the word). We'll then prepare our training and test data. We'll use a sliding window technique, adding one word at a time to each sentence.

<img src="LSTMExample.png" width="550">

```python
# Note: It is recommended that you print out the intermediate results for each of the steps below.
#       This can be very helpful towards understanding the operations in detail.

# Create a map from each word to its index.
word_to_index = dict((word, index) for index, word in enumerate(unique_words))

# Create a map from each index to its word.
index_to_word = dict((index, word) for index, word in enumerate(unique_words))

maxlen = 30

test_seed_sentences = []
test_predicted_words = []

# Cut the text in semi-redundant sequences of maxlen characters.
for line in lines:
    this_sentence = line.lower().strip().split(' ')
    for i in range(len(this_sentence) - 1):
        test_seed_sentences.append(this_sentence[:i + 1])
        test_predicted_words.append(this_sentence[i + 1])

# Let's see what this looks like to help build our intuition of how we will be training the model.
print(test_seed_sentences[0], test_predicted_words[0])
print(test_seed_sentences[1], test_predicted_words[1])
print(test_seed_sentences[2], test_predicted_words[2])

print('Number of sequences:', len(test_seed_sentences))
```

Next, we'll vectorize the inputs to a machine-understandable form.

<img src="VectorizePart1V3.png" width="700">
<img src="VectorizePart2.png" width="800">

```python
print('Vectorization...')

# Initialize the multi-dimensional vectors according to the above diagram. Initially, all slots will be zero-filled.
x = np.zeros((len(test_seed_sentences), maxlen, len(unique_words)), dtype=np.bool)
y = np.zeros((len(test_seed_sentences), len(unique_words)), dtype=np.bool)

# Fill in "1" values for according to the scheme depicted in the above diagram.
for sentence_index, sentence in enumerate(test_seed_sentences):
    for word_index, word in enumerate(sentence):
        x[sentence_index, word_index, word_to_index[word]] = 1
    y[sentence_index, word_to_index[test_predicted_words[sentence_index]]] = 1
    
# Let's visualize what this looks like. 
# x[1] represents the second sentence.
# y[1] represents the predicted word for the second sentence.

print(test_seed_sentences[1], test_predicted_words[1])

# We are going to be training the LSTM on sentences of a fixed length (30).
# Therefore, we expect the length of each x[n] be exactly 30. 
# x[n][0] represents the first word in the n-th sentence.
# The length of x[n][0] is equal to the number of possible unique words.
# There is a "True" value in x[i][j][k] if the k-th unique word is present in position j (of 30) for sentence i.
print(len(x[1]))
print('Index of word "alice":', word_to_index['alice'])
print('Index of word "was":', word_to_index['was'])
print('Index of word "beginning":', word_to_index['beginning'])
print(x[1][0])
print('"alice" should be the first word for the second sentence: ', x[1][0][word_to_index['alice']])
print('"was" should be the second word for the second sentence: ', x[1][1][word_to_index['was']])

# y[n] is structured similarly, except it holds only a single possible word for each n-th sentence.
# Therefore, we expect the length of each y[n] to be equal to the length of unique words.
print(len(y[1]))
print(y[1])
print('"beginning" should be the predicted word for the second sentence', y[1][word_to_index['beginning']])
```

Finally, we'll build our model. According to our vectorization scheme, each sentence entry in training set `x` points to an array of length `maxlen` (30), where each element points to another array of length `len(unique_words)`. Therefore, the shape of each individual sentence that will be input into the LSTM layer is `(maxlen, len(unique_words))`.

We'll use a sampling function to choose different predicted words based on a `temperature` parameter.

Running the following code block will start the training process. This process will take approximately 2 minutes to complete. Every 20 epochs, we'll generate a sentence using the model. We should expect to see the generated sentences get better over time. We should also expect the loss to decrease, and the accuracy to increase as training continues.

LSTMs (and RNNs in general) can be quite volatile, and may require a fair amount of training to achieve acceptable results. While we've capped the number of epochs to 100, we encourage you to train for many more epochs and observe how the text generation improves. Note that we're also using a very simple network architecture, and a very small amount of training data (small excerpt of the full *Alice in Wonderland* text).

**Congratulations!** You have successfully created your first machine learning-based text generator!

```python
print('Build model...')

# Construct a sequential model using the Keras API.
model = tf.keras.Sequential()

# The first layer will be an LSTM layer with 128 units. 
model.add(tf.keras.layers.LSTM(128, input_shape=(maxlen, len(unique_words))))

# The second layer will be a dense layer.
model.add(tf.keras.layers.Dense(len(unique_words)))
model.add(tf.keras.layers.Activation('softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Helper function to sample an index from a probability array.
def sample(preds, temperature=1.0):
    
    # Convert the predictions to the float64s.
    preds = np.asarray(preds).astype('float64')
    
    # Compute the logarithm of each prediction, and divide each by the provided "temperature".
    preds = np.log(preds) / temperature
    
    # Compute the exponential of all values.
    exp_preds = np.exp(preds)
    
    # Normalize the predictions based on the sum of the predictions.
    preds = exp_preds / np.sum(exp_preds)
    
    # Draw samples from a multinomial distribution.
    probas = np.random.multinomial(1, preds, 1)
    
    # Retrieve the answer based on the highest probability.
    return np.argmax(probas)


def predict():
    # We will invoke the sample function with a variety of different "temperatures"
    # This will produce a variety of different answers.
    for diversity in [0.2, 0.5, 1.0, 1.2]:
        
        print('')
        print('----- Diversity:', diversity)

        generated = []

        # Simple seed query.
        query = ["there"]

        generated.extend(query)
        sys.stdout.write(" ".join(generated))

        # Start from the seed query, predict the next word, and continously expand the seed query until
        # the maximum length (maxlen) amount of words are generated.
        for i in range(maxlen - len(query)):
            
            # Create the input multi-dimensional vector for this prediction step.
            x_pred = np.zeros((1, maxlen, len(unique_words)))
            
            # Prepare the input as per our encoding scheme.
            for word_index, word in enumerate(query):
                x_pred[0, word_index, word_to_index[word]] = 1.

            # Predict the next word.
            preds = model.predict(x_pred, verbose=0)[0]
            
            # Sample the next word.
            next_index = sample(preds, diversity)
            
            # Map the word index to the actual word.
            next_word = index_to_word[next_index]

            # Expand the set of generated words, and the query to be used for the next iteration.
            generated.extend([next_word])
            query.extend([next_word])

            # Print out the next word to the console.
            sys.stdout.write(" " + next_word)
            sys.stdout.flush()

        print()

class Metrics(tf.keras.callbacks.Callback):
    
    # Function invoked at end of each epoch.
    def on_epoch_end(self, epoch, logs={}):
        if epoch % 20 != 0:
            return

        # Every 20 epochs, we'll run the predict function and print out intermediate results.
        
        print()
        print('----- Generating text after Epoch: %d' % epoch)
        predict()


metrics = Metrics()

# Set random seed so we can reproduce similar results.
np.random.seed(0)

model.fit(x, y,
          batch_size=16,
          epochs=100,
          callbacks=[metrics])

predict()
```

```python
# Read the training set CSV file.
df_train = pd.read_csv(TRAIN_FILE)

# Perform an in-place replacement of categorical feature values to numerical feature values using 
# our custom feature mapping dictionary.
df_train.replace(feature_mapping, inplace=True)

# Read the test set CSV file.
df_test = pd.read_csv(TEST_FILE)

# Perform an in-place replacement of categorical feature values to numerical feature values using 
# our custom feature mapping dictionary.
df_test.replace(feature_mapping, inplace=True)

# Select all feature names.
features = df_train.columns[:-1]

# Select the Class column for both the training and test set.
train_true_y = df_train['Class']
test_true_y = df_test['Class']
```

Now that we've got our datasets loaded, let's train a random forest classifier, and evaluate it against the test set.

```python
# Train using a random forest classifier. We can control the number of trees that will be created using the
# n_estimators parameter.
clf = RandomForestClassifier(n_estimators=100)
clf.fit(df_train[features], train_true_y)

# Evaluate against training data.
train_pred_y = clf.predict(df_train[features])

print("Training Data Classification Report...")
print("")

# The classification_report function receives two inputs: The input features, and the expected outputs,
# and produces a precision/recall/f1-score/support report.
print(classification_report(train_true_y, train_pred_y, target_names=target_names))

# Map the expected outputs to human readable names.
labeled_train_true_y = [target_names[j] for j in train_true_y]

# Map the actual outputs to human readable names.
labeled_train_pred_y = [target_names[j] for j in train_pred_y]

# Create confusion matrix.
cm = confusion_matrix(labeled_train_true_y, labeled_train_pred_y, labels=target_names)

# The print_cm function is a utility function to print out a confusion matrix in a human-readable form.
print_cm(cm, target_names)

# Evaluate against test data.
test_pred_y = clf.predict(df_test[features])

print("")
print("Test Data Classification Report...")
print("")

# The classification_report function receives two inputs: The input features, and the expected outputs,
# and produces a precision/recall/f1-score/support report.
print(classification_report(test_true_y, test_pred_y, target_names=target_names))

labeled_test_true_y = [target_names[j] for j in test_true_y]
labeled_test_pred_y = [target_names[j] for j in test_pred_y]

cm = confusion_matrix(labeled_test_true_y, labeled_test_pred_y, target_names)

# The print_cm function is a utility function to print out a confusion matrix in a human-readable form.
print_cm(cm, target_names)
```

As we can see, the random forest classifier has achieved an f1-score of 1.00 against the training data. This is likely a sign that the training algorithm has overfit to the training data.
```python
# Train with a maximum depth classifier.
clf = RandomForestClassifier(n_estimators=100, max_depth=8)
clf.fit(df_train[features], train_true_y)

# Evaluate against training data.
train_pred_y = clf.predict(df_train[features])

print("Training Data Classification Report...")
print("")

print(classification_report(train_true_y, train_pred_y, target_names=target_names))
labeled_train_true_y = [target_names[j] for j in train_true_y]
labeled_train_pred_y = [target_names[j] for j in train_pred_y]
cm = confusion_matrix(labeled_train_true_y, labeled_train_pred_y, labels=target_names)
print_cm(cm, target_names)

# Evaluate against test data.
test_pred_y = clf.predict(df_test[features])

print("")
print("Test Data Classification Report...")
print("")

print(classification_report(test_true_y, test_pred_y, target_names=target_names))
labeled_test_true_y = [target_names[j] for j in test_true_y]
labeled_test_pred_y = [target_names[j] for j in test_pred_y]
cm = confusion_matrix(labeled_test_true_y, labeled_test_pred_y, target_names)
print_cm(cm, target_names)

print(target_names)
```

### Generating Abstract Test Cases

Much research exists on generating sequential information using ML-based techniques. In the field of natural language processing (NLP), several approaches to text generation have been studied. A conventional approach is to treat text generation as a sequence-to-sequence problem. A sequence of text is fed into a trainable algorithm that in turn outputs a sequence of text. The training goal is that the concatenation of both sequences results in a plausible sentence. Since machines do not understand words, a common practice is to map words to integer values, resulting in what we may refer to as a word encoding. This means that both the input and output of the sequence-to-sequence problem are a sequence of integers. Word encodings are challenging to create and maintain as there are many possible words that may appear, especially when different languages are used. 

An alternative approach is to create a character encoding, where each character (for example, each character from A-Z) is mapped to an integer. Although this simplifies the text-to-machine mapping process, it creates additional complexity from a training standpoint. There are many valid (and invalid) arrangements of character sequences that can first form words and eventually form sentences. In contrast, there are much fewer unique arrangements of complete words. ML-based text generation techniques that leverage character-based encoding typically require more training to reach stability and feasibility-of-use over word-based encoding approaches.

The question we will explore next is: How can text generation be leveraged to aid in software testing? 

In the previous section, we explored practical applications for perceiving web application state. Raising the level of abstraction by which we interact with objects in a web application sets the stage for the next step. Having the ability to recognize objects, the next task of interest is learning important test flows from human testers. A test flow is a sequence of actions performed onto a SUT, followed by a set of expected observations. The steps that a human tester follows while executing test flows may be modeled as follows:

<img src="TestFlowProcess.png" width="600">

At a high level, we present the test flow execution process as a sequence of steps where each step belongs to one of three categories: Perceive, Act, Observe. Steps belonging to the Perceive category focus on establishing preconditions for a test flow, and involve drawing observations from the environment, and action planning before selecting an appropriate set of actions. Steps belonging to the Act category focus on selecting and executing appropriate actions based on the previously constructed plan. Finally, steps belonging to the Observe category focus on comparing expected vs. actual SUT behavior, and deciding on correctness. We support the continuity property of the testing process by allowing the results of the Observe steps to initiate a new Perceive stage, thus forming a cycle. 

Presenting the test flow process using this framework is the first step towards the goal of being able to represent the process in a machine-understandable format. Next, we define a language that may be used to express concrete test flows that fit into our framework. The language must be expressive enough to allow covering essential test cases, yet constrained enough to reduce the data complexity of ML-based techniques to test flow generation. The language must also promote the use of webpage component abstractions that utilize the system described in Section 7.2 in place of SUT-specific information. This supports the generality of the approach.

The main building blocks of the language are components, actions, and observations. Components represent elements on a web page. Observations represent information about components that can be perceived from a given web page. Finally, actions are interactions that may be performed onto components. By interleaving observations and actions, the language allows for the specification of test flows. The language supports the use of abstract learned objects, instead of using specific input values and observed text values that may only be pertinent to a single SUT or only pertinent to a specific domain of software applications. To use test flows as training data, a sliding window data augmentation technique is used to expand the training data. We frame the test flow learning problem as a supervised learning problem by taking each test flow from the training set and expanding it one word at a time, creating a new training example with a more complete sentence at each iteration. For each generated training example, the next step of the sentence is the correct assigned label. 

Being able to train an ML system to generate test flows means we can now leverage humans not just to train a system to detect objects on a web page, but also to train a system on how those objects should interact in ways that are meaningful to testing.

Let's apply the techniques we learned in Chapter 6 (Neural Network Text Generation) to the problem of abstract test generation. We've supplied a small amount of training data in `data/seedAbstractTests.txt`. We'll process this file exactly like we did in the *Alice in Wonderland* example.

```python
import io
import random
import sys
import json

import numpy as np

# Open the seeded abstract tests file, and read the lines of the file.
with io.open('data/seedAbstractTests.txt', encoding='utf-8') as f:
    lines = f.readlines()

# Reconstruct the text by joining each of the lines, inserting a newline character in between each.
text = "\n".join(lines)
print('corpus length:', len(text))

# Convert the text into lowercase, and replace all new line characters with a single empty space.
text = text.lower().replace('\n', ' ')

# Split the string on all spaces. The result is a list of all the words.
words = text.strip().split(' ')

# Add the empty string to the list of words. We can use this string for padding purposes later on when we 
# prepare the input for the neural net algorithms.
words.append('')

# Convert the list of words to a set (to remove duplicates), and back to a list.
unique_words = list(set(words))
print('total unique words:', len(unique_words))

# Create a map from each word to its index.
word_to_index = dict((word, index) for index, word in enumerate(unique_words))

# Create a map from each index to its word.
index_to_word = dict((index, word) for index, word in enumerate(unique_words))

# Cut the text in semi-redundant sequences of maxlen characters
maxlen = 11

test_seed_sentences = []
test_predicted_words = []

for line in lines:
    this_sentence = line.lower().strip().split(' ')
    for i in range(len(this_sentence) - 1):
        test_seed_sentences.append(this_sentence[:i + 1])
        test_predicted_words.append(this_sentence[i + 1])

print('Number of sequences:', len(test_seed_sentences))

print('Vectorization...')

# Initialize the multi-dimensional vectors according to our encoding scheme (see Section 6). 
# Initially, all slots will be zero-filled.
x = np.zeros((len(test_seed_sentences), maxlen, len(unique_words)), dtype=np.bool)
y = np.zeros((len(test_seed_sentences), len(unique_words)), dtype=np.bool)

# Fill in "1" values for according to the encoding scheme.
for sentence_index, sentence in enumerate(test_seed_sentences):
    for word_index, word in enumerate(sentence):
        x[sentence_index, word_index, word_to_index[word]] = 1
    y[sentence_index, word_to_index[test_predicted_words[sentence_index]]] = 1
    
print('Build model...')

# Construct a sequential model using the Keras API.
model = tf.keras.models.Sequential()

# The first layer will be an LSTM layer with 128 units. 
model.add(tf.keras.layers.LSTM(128, input_shape=(maxlen, len(unique_words))))

# The second layer will be a dense layer.
model.add(tf.keras.layers.Dense(len(unique_words)))
model.add(tf.keras.layers.Activation('softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Helper function to sample an index from a probability array.
def sample(preds, temperature=1.0):
    
    # Convert the predictions to the float64s.
    preds = np.asarray(preds).astype('float64')
    
    # Compute the logarithm of each prediction, and divide each by the provided "temperature".
    preds = np.log(preds) / temperature
    
    # Compute the exponential of all values.
    exp_preds = np.exp(preds)
    
    # Normalize the predictions based on the sum of the predictions.
    preds = exp_preds / np.sum(exp_preds)
    
    # Draw samples from a multinomial distribution.
    probas = np.random.multinomial(1, preds, 1)
    
    # Retrieve the answer based on the highest probability.
    return np.argmax(probas)


def predict(input_query=None):
    # We will invoke the sample function with a variety of different "temperatures"
    # This will produce a variety of different answers.
    for diversity in [0.2, 0.5, 1.0, 1.2]:
        print('')
        print('----- Diversity:', diversity)

        generated = []

        # Simple seed query.
        query = ["observe", "required", "textbox", "firstname", "try"]
        
        if input_query is not None:
            query = input_query

        generated.extend(query)
        sys.stdout.write(" ".join(generated))

        # Start from the seed query, predict the next word, and continously expand the seed query until
        # the maximum length (maxlen) amount of words are generated.
        for i in range(maxlen - len(query)):
            
            # Create the input multi-dimensional vector for this prediction step.
            x_pred = np.zeros((1, maxlen, len(unique_words)))
            
            # Prepare the input as per our encoding scheme.
            for word_index, word in enumerate(query):
                x_pred[0, word_index, word_to_index[word]] = 1.

            # Predict the next word.
            preds = model.predict(x_pred, verbose=0)[0]
            
            # Sample the next word.
            next_index = sample(preds, diversity)
            
            # Map the word index to the actual word.
            next_word = index_to_word[next_index]

            # Expand the set of generated words, and the query to be used for the next iteration.
            generated.extend([next_word])
            query.extend([next_word])

            # Print out the next word to the console.
            sys.stdout.write(" " + next_word)
            sys.stdout.flush()

        print()

class Metrics(tf.keras.callbacks.Callback):
    
    # Function invoked at end of each epoch.
    def on_epoch_end(self, epoch, logs={}):
        if epoch % 20 != 0:
            return

        # Every 20 epochs, we'll run the predict function and print out intermediate results.
        
        print()
        print('----- Generating text after Epoch: %d' % epoch)
        predict()


metrics = Metrics()

# Set random seed so we can reproduce similar results.
np.random.seed(0)

model.fit(x, y,
          batch_size=8,
          epochs=100,
          callbacks=[metrics])

predict()
```
